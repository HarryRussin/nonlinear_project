# 1. Introduction
Systems that oscillate are generally referred to as harmonic. In physics, this specifically refers to systems where a restoring force is proportional to the displacement experienced by the system. For such simple systems, we can predict a regular periodic motion. Some simple harmonic systems, like a mass on a spring, can be modelled by ordinary differential equations (ODEs) which we would describe as linear equations. Other systems, like the pendulum, are described by a nonlinear ODE due to the presence of trigonometric functions.  Conversely, some oscillating systems do not exhibit periodic motion but instead oscillate with a variety of different modes and periods, frequencies. A 'simple' example is the driven damped pendulum. These systems are referred to as being anharmonic and can exhibit complex behaviour such as chaos.

Physics courses are mostly concerned with exposing students to understand and solve linear systems. Even when dealing with [nonlinear systems](https://en.wikipedia.org/wiki/Nonlinear_system#Nonlinear_differential_equations) students are taught how to simplify by removing the nonlinear component. The typical example of this is the small angle approximation in simplifying the pendulum. Unfortunately, removing the nonlinear behaviour also removes much of the rich physics. Many real world systems are interesting because they are nonlinear, so it is a shame that students do not get exposed to useful methods to solve and appreciate the complexity of such systems. These methods are computational and by using simple algorithms students with basic coding skills can directly solve and begin to explore the nonlinear world.

Many of the algorithms used in computational methods have existed before modern computers. Examples include the Euler and Runge-Kutta methods developed in the 17th and 19th centuries respectively.  Coupled with the processing power of modern computers, the combination has enabled nonlinear dynamics to emerge as a new active field over the past few decades. The classic text book [Nonlinear Dynamics and Chaos](https://find.shef.ac.uk/permalink/f/1lephdb/44SFD_ALMA_DS51319938900001441) by Strogatz is a great introduction to this field [1]. You may already be familiar with several nonlinear applications in physics such as weather behaviour (Lorentz attractor, butterfly effect), fluid turbulence (Navier Stokes). But this field also sees applications to many other disciplines including biology (population growth, synchronisation), economics (markets) and mathematics (fractals, complexity).

# 2. Theory
Recurrence describes how a system returns to similar states over time. The simplest case of recurrence is periodic motion which follows a stable repeating cycle such as in simple harmonic motion (SHM). Real-world systems however often exhibit effects like damping and driving forces. These lead to more complex recurrence, including quasi-periodicity and chaos. To model such behaviour, we can use ordinary differential equations (ODEs) to describe how the physics of the system changes with time. Here in this project we will start with a simple harmonic system and then add more nonlinear effects in successive systems. These theoretical models in the form of ODEs can then be utilised in our computational methods in Section 3.

An example of simple harmonic motion is the mass-spring system. This can be modelled via Hooke's law which describes the restoring force $F$ that acts on the mass $m$ from the spring constant $k$:
$$
    F=-k x
$$
As the spring is compressed ($x<0$) or stretched ($x>0$) the restoring force is a linear function of the mass's position $x$. We can rewrite the equation in terms of Newton's 2nd lab
$$
    ma =-k x
$$
and again in terms of the second derivative of position
$$
    m \frac{d^2 x}{dt^2}=-k x
$$
and finally arrive at the ODE to model this mass-spring system:
$$
   \frac{d^2 x}{dt^2}=\frac{-k}{m} x
$$

Note that this is a second order ODE. Later when we use our numerical methods we will simply write this ODE as two first order equations for velocity $$\frac{dx}{dt} = v$$ and acceleration $$\frac{dv}{dt} = a = \frac{-k}{m} x $$

As physicists we are interested in determining the solution to such ODEs, which in this case is the function $x(t)$.  The analytical solution for this simple mass-spring ODE is
$$
x(t) = A \cos{\sqrt{\frac{k}{m}} t}
$$
As computational physicists we start with this simple linear example because we can solve the ODE for an analytical solution. This enables us to test the accuracy of our computational methods (see Section 3) before their application to more complex systems.

As we move towards modelling more complex systems, we will see nonlinear relationships within their ODEs. For example, the next harmonic oscillator on our short project will be the pendulum. Here the position of a mass is defined as the angle $\theta$ of an attached string with length $l$. The equilibrium position is the mass suspended straight down where $\theta=0$. The angular coordinate is negative for left of equilibrium and positive for right of equilibrium.  As the angle increases from equilibrium the restoring force is provided by gravity  $$  \frac{d^2 \theta}{dt^2} = - \frac{g}{\ell} \sin \theta.$$ Instead of being a linear function of position $\theta$, the presence of $\sin \theta$ implies a nonlinear relationship. Can you [solve this analytically](https://scholar.google.co.uk/scholar?hl=en&as_sdt=0,5&q=analytical+solution+of+pendulum) to predict the solution $\theta (t)$ for any initial angle? Good luck! When we use computational methods, we will decompose such 2nd order ODEs to two first order equations:
$$   \frac {d \theta}{dt} =  \omega \\
    \frac {d \omega}{dt}  = - \frac{g}{\ell} \sin \theta
$$

A nice feature of ODE models is that we can add more physics to the systems quite easily. If we wish to include the effects of damping for example, we can include a term which scales with velocity and some dampening factor $\beta$
$$ \frac{d^2 \theta}{dt^2} = - \frac{g}{\ell} \sin \theta - \beta\frac{d \theta}{dt}
$$
And if we wish to provide a driving force to push/pull the pendulum we can add a positive acceleration via an external periodic force with amplitude $A$ that oscillates with a driver frequency $\omega_D$
$$   \frac{d^2 \theta}{dt^2} = - \frac{g}{\ell} \sin \theta - \beta\frac{d \theta}{dt}  + A\cos(\omega_D t)
$$
The combination of the driver force, dampening and the restoring force will create complex physics that is very difficult (impossible?) to solve analytically. Once we build our computational model, we can solve this ODE numerically and then investigate nonlinear behaviour including chaos. Now that we have defined the physical models via these ODEs, let us move to begin our project by building and benchmarking our computational methods.


# 3. Computational Methods

During the first week of this course, we developed numerical methods to solve ODEs. These were the Runge-Kutta methods which essentially estimate how the solution for an ODE changes over some short interval. For definition purposes, let us assume the solution to some ODE is $y(x)$. Unlike analytical methods, we will only solve the ODE in discrete steps. We choose some small step size $h$ and also choose the initial value of the solution $y(0)$. The Runge-Kutta methods calculate an approximate change in the solution ($k$=\Delta y) by evaluating its derivative function $f(x,y)$.  The standard method for this project will be the RK4 method. This evaluates the derivative function (i.e. the ODE) at 4 points over the small step $h$ and produces 4 estimates of the change in the ODE solution. The standard RK4 algorithm is:

$$ k_1 = h\times f(x,y) $$
$$ k_2 = h\times f(x + \frac{h}{2}, y + \frac{k_1}{2}) $$
$$ k_3 = h\times f(x + \frac{h}{2}, y + \frac{k_2}{2}) $$
$$ k_4 = h\times f(x + h, y + k_3) $$

By now you should have noticed how the ODE $f(x,y)$ is evaluated at different points from the starting value $x$, to the midpoint $x+h/2$, and finally the end point $x+h$. The 4 k-values are then averaged with double weighting given to the more accurate min point estimates. The value of the solution at the end of the small step is then:

$$
y = y + \frac{k_1 + 2k_2 + 2k_3 + k_4}{6}
$$

When we combine this method with a reiterative for-loop, we can repetitively perform the calculation where the new value of $y$ at the end of a step becomes the initial value for the next step and so on for some arbitrary number of steps. For us physicists, the independent variable $x$ is time so the step size $h$ is the time step in our simulations.
